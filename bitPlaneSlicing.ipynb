```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
im=cv2.imread('baboon.tiff',0)
plt.imshow(im,'gray')
plt.title("Baboon")
plt.show()
```

```python
im[:10,:10]
```

```python
bit=[]
for i in range(8):
    bitplane=im&(1<<i)
    bit.append(bitplane)
fig=plt.figure(figsize=(8,8))
for i in range(8):
    plt.subplot(2,4,i+1)
    plt.title('Bit plane img '+str(i))
    plt.imshow(bit[i],'gray')
plt.show()
```
