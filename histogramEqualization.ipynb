# Histogram

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import math
import os
```

## Histogram Plotting

```python
image=cv2.imread("bee.tiff",0)
```

```python
plt.imshow(image,cmap='gray')
plt.show()
```

```python
def plotHist(image):
    imHist = image.flatten()
    imHist = list(imHist)
    gScale = [x for x in range(256)]
    freq = []
    for i in gScale:
        freq.append(imHist.count(i))
    plt.plot(gScale,freq)
    plt.show()
```

```python
plotHist(image)
```

# Histogram Equalization

```python
image = cv2.imread("bee.tiff",0)
```

```python
plt.imshow(image,cmap = 'gray')
plt.show()
```

```python
imHist2 = image.flatten()
```

```python
imHist2 = list(imHist2)
```

```python
# print(imHist)
```

```python
# for element in imHist:
    # print(element)
```

```python
gScale2 = [x for x in range(256)]
```

```python
freq2 = []
for i in gScale2:
    freq2.append(imHist2.count(i))
```

```python
nHist2 = []
for x in freq2:
    nHist2.append(x/image.size)
```

```python
cProb = np.cumsum(np.array(nHist2))
```

```python
sk=[]
for x in cProb:
    sk.append(round(255*x))   #use to map the original pixel intensities to new values during equilization
```

```python
newImage = np.zeros((512,512))
```

```python
for row in range(512):
    for col in range(512):
        newImage[row][col] = sk[image[row][col]]
```

```python
plt.imshow(image,cmap='gray')
plt.show()
```

```python
plt.imshow(newImage,cmap='gray')
plt.show()
```

```python
plt.plot(gScale2,sk)
plt.show()
```

```python
plotHist(newImage)
```
